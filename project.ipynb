1.	INTRODUCTION

a.	Problem Description

Football market is a huge industry which may be profitable not only for those actively taking part in a football game but also for those who just watch it. In this end, a huge brother football industry that of bookmarking has developed. Although bookmarkers try to provide odds that should lead to their profitability there are also attempts based on their predictors that one may use to overcome the first. However, bookmarker odds are not the only factors which may be considered to predict the results of a game. Other performance metrics such as ball possession or predictors introduced based on past activity of the teams may be examined. 
In this regard, the object of this project is to develop a model using machine learning methods (Lasso regression, Decision tree and their ensemble), based on which future match outcomes can be predicted. The results are provided as probabilities for each of the three outcomes (home win, draw and away win). Features are extracted based on past data provided and the time dependency is considered during feature extraction.As a measure of wellness of the developed approach predictions for 8 rounds for the Premier League are aimed to be given. Firstly, appropriate features are extracted by the provided data set, then the data is preprocessed and finally the machine learning approach is selected and its parameters are tuned.

b.	Summary of the Proposed Approach

As it is already mentioned, the outermost object of the study is to provide probabilities for the possible outcomes of a game. Firstly, the probable outcomes of a game were numerically assigned. However, the predicted value was not handled as a categorical but rather as a regressed variable not to omit its grading nature. Then, the predicted values of the regression have been transformed into categories based on predetermined intervals. Feature space is constructed using odds, number of points earned by home team and away team in the last 5 matches, average goals scored by home team and away team in the last 5 matches. Data is partitioned into two parts: training set and test set. Models are constructed with training data using lasso, decision trees and kNN + lasso. Results are checked by using test set and misclassification error and RPS are calculated to interpret results.

2.	RELATED LITERATURE

An early attempt to predict the result of a sports game is that of Wood (1945) trying to model the performance in a cricket match using the coefficient of variation. Additionally, predicting the outcome of a football match has its roots back to 1956 where Moroney tried to predict the number of goals scored using Poisson distribution and negative binomial distribution. Following, Reep et al. (1971) tried to apply the same distributions in other sport branches (cricket, ice-hokey, baseball, lawn tennis), too, based on on-time performances. 
Nowadays, the attempts to improve the performance of the methods predicting soccer games outcome continue. With the emergence of increased data collection and data analysis methods the approaches are also diversified. An example can be that of  Kaunitz et al. (2017) who tried to employ a strategy, as they call it, which uses the odd information provided by the bookmarkers designed in favor of them, indicating the mispredicted games. It is noted, that their philosophy was profitable in the long run (simulated 10-year period). Moreover, Cho et al. (2018) have used gradient boosting and social network analysis approaches to predict the win-lose result. Among the compared methods /decision tree, neural network, Support vector machines (SVM) and others) it is underlined that gradient boosting gives the best result. Gu et al. (2019) have also preferred SVM but combined it with principal component analysis, and nonparametric statistical analysis. It is asserted that a prediction accuracy of more than 90% has been achieved. However, these are only some examples of the vast literature on the topic and the interest on the field still increases. 

3.	APPROACH

The main target of the model construction process is to find a successful relation between results of soccer matches and the information we have about that match. Result of a soccer match is three class categorical variable. These categorical variables are transformed into numerical variables. More specifically 0, 0.5 and 1 were assigned to home win, draw and away win respectively.  Then the interval of the obtained outcomes (0 to 1) was divided into 3 parts. The results of the game were assigned based on the proximity to assigned values. The probabilities of the outcome on the other hand was calculated as the ratio of the exponentials of the variation from each value to the sum of the variations. The reason that exponential was used is to provide milder probabilities. The next step was to select the features to be used during model construction. In this context, the odds(odd_1, odd_x and odd_2) provided by the bookmarkers were primarily examined. It may be pointed out that there are also studies implying the odds provided may be mispriced and biased (Franke M.).  Then the total number of points collected by each team (home and away) are calculated. It should be hereby noted that the time dependent nature of the problem was taken into account. In other words, when calculating the points for each team the points till the specific match were considered. Furthermore, the fact that the teams played home or away is also considered by introducing two different predictors for each team: one neglecting and the other considering the match place. Similarly, four other predictors averaging the till then last five game number of goals are used. 
After constructing the dataset, the observations with NA values were disregarded. These missing values were either because no odd information was provided or because less than five games till that game have conducted. No replacement with mean or median has been chosen because the number of data points with NA values was high so their replacement may mislead the results. Additionally, the total set was divided into two: 2400 points were used for training and 500 for testing the models obtained. 
The next step consists of the model building. Two methods and a combination of them are used. Firstly, kNN + lasso penalized regression was selected with kNN and lambda chosen so as to minimize the cv error. Then a decision tree was constructed with a cp parameter so as to minimize the misclassification error in the test set. Finally, the probabilities of the two methods were averaged leading to an ensemble prediction of the to methods.

4.	RESULTS & DISCUSSION

Results are evaluated by using two methods: Average RPS and misclassification error. Misclassification error is determined by dividing number of incorrect guesses to total number of matches, in other words, length of the test set. The significant part of this process is to assign numerical regression result to its category. As it is mentioned in the approach part; 0, 0.5, and 1 are assigned to home win, tie, and away win respectively. If the predicted value is smaller than 0.4 it is assigned to home win, and if it is larger than 0.6 it is assigned to away win, and other results are assigned to tie. The interval can be divided into 3 equal parts but there is a problem in that case. When it is divided into 3 equal parts, most of the games are predicted as draw because tie is in the middle of the interval. Hence, the result tie takes advantage of this situation. The interval of the tie is shortened for that reason. 
RPS (Ranked Probability Score) is also calculated to evaluate results. Transforming numerical result into 3 probabilities is also tricky process. Distances between the predicted result and 0, 0.5 and 1 are considered most meaningful information for this transformation. 



Results show that decision tree is the best model according to both average RPS and misclassification error.  Each of the models has a misclassification error lower than 0.5. This is a good sign but none of them is reliable for making bets. 

5.	CONCLUSIONS & FUTURE WORK

In this project, it is aimed to build a successful model to predict results of soccer matches. k-nearest neighbor + lasso and decision tree are the methods selected to build models. Ensemble model is also constructed by taking average of predictions made by lasso and decision tree. It is considered as more robust method. As a result, decision tree is the best model according the prediction results. 

These models may be improved by expanding feature space by searching more meaningful predictors.
